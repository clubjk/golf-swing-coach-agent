```python
# This cell loads all the necessary Python packages and modules required to work with the OpenAI Agents SDK and MCP (Model Context Protocol) servers.

# The imports
from dotenv import load_dotenv          # Loads environment variables from the .env file into the Python environment
from agents import Agent, Runner, trace # Imports the core Agent class, Runner for execution, and trace for observability
from agents.mcp import MCPServerStdio   # Imports the MCPServerStdio class used to spawn and communicate with MCP servers
import os                               # Imports the os module to work with file paths and environment
```

```python
# This cell loads environment variables (such as API keys) from the .env file so they are available to the application.

load_dotenv(override=True)  # Loads variables from .env file and overrides any existing environment variables with the same names
```

### Let's use MCP in OpenAI Agents SDK

1. Create a Client

2. Have it spawn a server

3. Collect the tools that the server can use

Let's try the Fetch mcp-server that we looked at last week

### Special note - in case you updated the OpenAI Agents framework:

OpenAI Agents SDK recently upgraded their SDK and made a breaking change (thank you, OpenAI 😂). I haven't yet updated the version in this repo to the latest version. If you've upgraded the package yourself, then you might get an error about missing positional arguments. If you get that error, you'll need to replace `await server.list_tools()` with `await server.session.list_tools()` and change the last line from `fetch_tools` to `fetch_tools.tools`. And you'll need to make this change every time you get this error! The joys of working at the bleeding edge..

```python
# This cell defines parameters for the fetch MCP server and demonstrates how to start it using MCPServerStdio, retrieve its available tools, and display them.

fetch_params = {"command": "uvx", "args": ["mcp-server-fetch"]}  # Creates a dictionary specifying the command and arguments needed to launch the mcp-server-fetch tool via uvx

async with MCPServerStdio(params=fetch_params, client_session_timeout_seconds=60) as server:  # Starts the MCP server using stdio transport with a 60-second timeout for the client session
    fetch_tools = await server.list_tools()  # Asynchronously calls the server to retrieve the list of tools it exposes

fetch_tools  # Displays the list of available tools returned by the fetch MCP server
```

## And now 2 more MCP servers - this time Javascript based, with node

### If the next cell hangs, see the important troubleshooting at the end of the SETUP-node instructions linked above.

```python

# This cell defines parameters for the Playwright MCP browser server and demonstrates starting it to retrieve its available tools.

playwright_params = {"command": "npx","args": [ "@playwright/mcp@latest"]}  # Creates configuration dictionary to launch the official Playwright MCP server using npx

async with MCPServerStdio(params=playwright_params, client_session_timeout_seconds=60) as server:  # Launches the Playwright MCP server with a 60 second client session timeout
    playwright_tools = await server.list_tools()  # Asynchronously fetches the list of tools provided by the Playwright MCP server

playwright_tools  # Outputs the list of browser automation tools made available by the server
```

```python

# This cell sets up the filesystem MCP server by creating a sandbox directory path and launching the server to list its file operation tools.

sandbox_path = os.path.abspath(os.path.join(os.getcwd(), "sandbox"))  # Creates an absolute path to a 'sandbox' folder in the current working directory
files_params = {"command": "npx", "args": ["-y", "@modelcontextprotocol/server-filesystem", sandbox_path]}  # Defines parameters to run the official filesystem MCP server pointed at the sandbox directory

async with MCPServerStdio(params=files_params,client_session_timeout_seconds=60) as server:  # Starts the filesystem MCP server using the defined parameters with a 60-second timeout
    file_tools = await server.list_tools()  # Asynchronously retrieves the list of file system tools (read, write, etc.) provided by this server

file_tools  # Displays the available filesystem tools from the MCP server
```

### And now.. bring on the Agent with Tools!

```python
# This cell creates a detailed system prompt for an investigative agent and then runs it with two MCP servers (filesystem + browser) to complete a real-world research task.

instructions = """
You are an expert recipe researcher. Your goal is to find a high-quality Shepherd's Pie (also spelled Shepherds Pie or Cottage Pie) recipe from a reputable site.

**Workflow:**
1. Use browser tools to search Google or directly navigate to recipe sites like bbcgoodfood.com, allrecipes.com, or deliciousmagazine.co.uk.
Be smart with the use of search terms; for example, "Shepherd's Pie recipe" or "Cottage Pie recipe" or "Shepherds Pie recipe" or "Shepherd Pie recipe"
Search 5 recipes and pick what you think is the best one.
2. Accept cookie consents quickly (click "Accept all", "Allow all", or "I agree" buttons).
3. Find a detailed recipe with ingredients and step-by-step instructions.
4. Extract the key information cleanly (title, ingredients list, method steps, servings, time).
5. Create a well-formatted Markdown summary.
6. Use the write_file tool to save it as "shepherd_pie.md" in the current sandbox directory.
7. Be efficient - avoid getting stuck in loops. If one approach fails, try a different search or site immediately.
8. Once the file is written, confirm success.

Focus on completing the task in as few turns as possible.
"""  # Enhanced instructions for more efficient recipe research and file writing

async with MCPServerStdio(params=files_params, client_session_timeout_seconds=120) as mcp_server_files:  # Starts the filesystem MCP server so the agent can read/write files
    async with MCPServerStdio(params=playwright_params, client_session_timeout_seconds=120) as mcp_server_browser:  # Starts the Playwright browser MCP server so the agent can browse the internet
        agent = Agent(  # Creates a new Agent instance configured with name, instructions, model, and the two MCP servers
            name="investigator", 
            instructions=instructions, 
            model="gpt-4o-mini",  # Updated to a current, reliable OpenAI model
            mcp_servers=[mcp_server_files, mcp_server_browser]
            )
        with trace("investigate"):  # Begins an OpenAI trace context to record and observe the agent's execution for debugging
            result = await Runner.run(
                agent, 
                "Find a great recipe for Shepherd's Pie, then summarize it in clean markdown format and save it to shepherd_pie.md in the sandbox directory.",
                max_turns=25  # Increased from default 10 to allow for browser navigation, cookie handling, content extraction and file writing
            )  # Runs the agent asynchronously with a specific task
            print(result.final_output)  # Prints the final output from the agent after it completes its task
```
